```python
# Which age certifications are the most popular for each genre?
# More adult oriented movies are generally the most popular, with notable exceptions being the 'animation' and 'family' category

genres = titles_df.iloc[:,15:].columns
ratings = np.delete(titles_df['age_certification'].unique(),6)

pop_means = pd.DataFrame(columns=ratings, index=genres)
rating_means = pd.DataFrame(columns=ratings, index=genres)

for genre in genres:
    for rating in ratings:
        pop_means.loc[genre,rating]=titles_df[(titles_df[genre]==1) & (titles_df['age_certification']==rating)]['tmdb_popularity'].mean()
        rating_means.loc[genre,rating]=titles_df[(titles_df[genre]==1) & (titles_df['age_certification']==rating)]['tmdb_score'].mean()
        if (math.isnan(pop_means.loc[genre,rating])):
            pop_means.loc[genre,rating]=0
        if (math.isnan(rating_means.loc[genre,rating])):
            rating_means.loc[genre,rating]=0
            
ax = pop_means.loc[:,['G','PG','PG-13','R']].plot.bar(figsize=(20,5))
plt.xlabel('Genre')
plt.ylabel('Average Rating')
```

```python
# On the other hand, rating is relatively constant across genres and maturity ratings
ax = rating_means.loc[:,['G','PG','PG-13','R']].plot.bar(figsize=(20,5))
plt.xlabel('Genre')
plt.ylabel('Average Rating')
```
